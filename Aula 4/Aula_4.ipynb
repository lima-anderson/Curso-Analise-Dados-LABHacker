```
#agregar dados por município e candidato
# mantem as colunas entre chaves e pede para somar o resto, ou seja, a colunca que sobrou, QT_VOTOS_NOMINAIS
df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO',  'NM_URNA_CANDIDATO']).sum()
```

```
# transformando novamente em dataframe, pq estava noutro formato

df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO',  'NM_URNA_CANDIDATO']).sum().reset_index()

df_votacao.sample(4)
```

```
#ordenar
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)

#excluir linhas duplicadas, manter a primeira
df_votacao.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)


df_votacao.head(10)
```

```
df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE']
```

```
#criação de coluna para representar o resultado da eleição no município
#true para bolsonaro, false para haddad

df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```
#https://github.com/estadao
#recuperar correspondencia

link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

df_equivalencia = pd.read_csv(link)

df_equivalencia.head()
```

```
#reindexação do dataframe de votação  pelo código do tse

df_voto_novo = df_votacao.copy()

df_voto_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)

df_voto_novo.head()
```

```
#reindexação do dataframe de equivalencia  pelo código do tse

df_equi_novo = df_equivalencia.copy()

df_equi_novo.set_index('COD_TSE', drop=False, inplace=True)

df_equi_novo.head()
```

```
#verifcando se está mesmo com os mesmos id's 
df_equi_novo[df_equi_novo['NOME'] == 'SÃO PAULO']
```

```
df_equi_novo.info()
```

```
df_voto_novo.info()
```

```
# ambos tem 5570 registros

# criação de novo dataframe com base no merge de votação  e equivalência

df_voto_equi = df_voto_novo.join(df_equi_novo)

df_voto_equi.head()
```

```
# unificar o dataframe do geopandas ao dataframe df_voto_equi
df_voto_equi.info()
```

```
df_mapa.info()
```

```
# transformar tipo de dado da coluna GEOCOD_IBGE
df_voto_equi['GEOCOD_IBGE'] = df_voto_equi['GEOCOD_IBGE'].astype(str)

df_voto_equi.info()
```

```
#unificação de indices
#modificar os índices do mapa

df_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)
```

```
# modificar os índices do dataframe df_voto_equi
df_voto_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)
```

```
#unificar o dataframe df_voto_equi com o dataframe de mapa
df_mapa_novo = df_mapa.join(df_voto_equi)

df_mapa_novo.head().T
```

```
df_mapa_novo.info()
```

```
# "municípios a mais" 
df_mapa_novo[df_mapa_novo['NR_CANDIDATO'].isnull()]
```

```
#eliminar informações acima
df_mapa_novo.dropna(inplace=True)
```

```
#referência ao código de cor, False/0 = red, True/1 = green
from matplotlib.colors import  ListedColormap

cmap = ListedColormap(['red', 'green'])

fig, ax = plt.subplots(1, figsize=(12,12))

ax = df_mapa_novo.plot(column='RESULTADO', cmap=cmap, legend=True, linewidth=1, ax=ax, edgecolor='grey')

ax.set_axis_off()

#título
ax.set_title('Resultado Eleição 2018 por Município / 2º Turno', fontdict={'fontsize': '25', 'fontweight': '3'})

#nota de rodapé
ax.annotate('Fonte: IBGE, 2015; TSE, 2019; Jornal Estado de São Paulo, 2019', xy=(0.1, 0.08), xycoords='figure fraction',
            horizontalalignment='left', verticalalignment='top', fontsize=10, color='#555555')

#alterando leganda
leg = ax.get_legend()
leg.get_texts()[0].set_text('Fernando Haddad')
leg.get_texts()[1].set_text('Jair')

plt.show()
```
